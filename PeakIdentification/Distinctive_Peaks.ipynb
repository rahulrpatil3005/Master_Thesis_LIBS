```python
import os
import pandas as pd

def search_wavelength_in_csv(folder_path, wavelength, tolerance=0.001):
    # List to store names of files containing the wavelength within the tolerance range and the detected wavelengths
    results = []

    # Traverse through all files in the specified folder
    for file_name in os.listdir(folder_path):
        # Check if the file is a .csv file
        if file_name.endswith('.csv'):
            file_path = os.path.join(folder_path, file_name)
            try:
                # Read the CSV file into a DataFrame
                df = pd.read_csv(file_path)

                # Find the wavelengths within the tolerance range
                mask = abs(df['obs_wl_air(nm)'] - wavelength) <= tolerance
                detected_wavelengths = df['obs_wl_air(nm)'][mask].tolist()

                if detected_wavelengths:
                    results.append((file_name, detected_wavelengths))
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

    return results

# Usage
folder_path = 'NIST_3'  # Replace with the path to your folder containing the CSV files
wavelength = 293.93 # Replace with the wavelength you want to search for
tolerance = 0.04  # Set the tolerance for the wavelength search

results = search_wavelength_in_csv(folder_path, wavelength, tolerance)

if results:
    print(f"The wavelength {wavelength} nm (within a tolerance of {tolerance} nm) is found in the following files:")
    for file_name, detected_wavelengths in results:
                print(f"File: {file_name}, Detected Wavelengths: {detected_wavelengths}")
else:
    print(f"The wavelength {wavelength} nm (within a tolerance of {tolerance} nm) is not found in any files.")
```
